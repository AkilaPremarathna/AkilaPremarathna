# DATA PREPROCESSING

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

def safe_numeric_convert(series):
    """Convert series to numeric, handling various string formats"""
    series = series.astype(str).str.replace(',', '')  # Remove commas
    series = series.str.replace('-', '0')  # Replace dashes with 0
    series = series.str.replace('nil', '0', case=False)  # Replace nil with 0
    series = series.str.replace('n/a', '0', case=False)  # Replace n/a with 0
    series = series.str.replace('N/A', '0', case=False)  # Replace N/A with 0
    series = series.str.replace('', '0')  # Replace empty strings with 0
    series = series.str.strip()  # Remove whitespace
    return pd.to_numeric(series, errors='coerce').fillna(0)

print("="*60)
print("DATA PREPROCESSING FOR AGRICULTURAL DATASET")
print("="*60)

# Create a copy of the original dataframe
df_preprocessed = df.copy()

print(f"Original dataset shape: {df_preprocessed.shape}")
print(f"Number of districts: {df_preprocessed['District'].nunique()}")

# Get all production and area columns
production_cols = [col for col in df_preprocessed.columns if 'PROD' in col and 'M.T.' in col]
area_cols = [col for col in df_preprocessed.columns if 'AREA' in col and 'Hect.' in col]

print(f"Found {len(production_cols)} production columns")
print(f"Found {len(area_cols)} area columns")

# Convert all production columns to numeric
print("\nConverting production columns to numeric...")
for col in production_cols:
    original_dtype = df_preprocessed[col].dtype
    df_preprocessed[col] = safe_numeric_convert(df_preprocessed[col])
    print(f"  {col}: {original_dtype}  numeric")

# Convert all area columns to numeric
print("\nConverting area columns to numeric...")
for col in area_cols:
    original_dtype = df_preprocessed[col].dtype
    df_preprocessed[col] = safe_numeric_convert(df_preprocessed[col])
    print(f"  {col}: {original_dtype}  numeric")

# Add total production per district
df_preprocessed['Total_Production_MT'] = df_preprocessed[production_cols].sum(axis=1)

# Add total area per district
df_preprocessed['Total_Area_Hectares'] = df_preprocessed[area_cols].sum(axis=1)

# Add overall yield efficiency per district
df_preprocessed['Overall_Yield_Efficiency'] = np.where(
    df_preprocessed['Total_Area_Hectares'] > 0,
    df_preprocessed['Total_Production_MT'] / df_preprocessed['Total_Area_Hectares'],
    0
)

print("\n" + "="*60)
print("PREPROCESSING COMPLETE!")
print("="*60)

print(f"Final dataset shape: {df_preprocessed.shape}")
print(f"New columns added:")
print(f"  - Total_Production_MT: Total production per district")
print(f"  - Total_Area_Hectares: Total cultivated area per district")
print(f"  - Overall_Yield_Efficiency: Production per hectare per district")

print(f"\nData quality summary:")
print(f"  - All production columns converted to numeric")
print(f"  - All area columns converted to numeric")
print(f"  - Missing/invalid values replaced with 0")
print(f"  - Ready for analysis!")

print(f"\nTop 5 districts by total production:")
top_districts = df_preprocessed.nlargest(5, 'Total_Production_MT')[['District', 'Total_Production_MT']]
for idx, row in top_districts.iterrows():
    print(f"  {row['District']}: {row['Total_Production_MT']:.1f} MT")

print(f"\nTop 5 most efficient districts (yield per hectare):")
top_efficient = df_preprocessed.nlargest(5, 'Overall_Yield_Efficiency')[['District', 'Overall_Yield_Efficiency']]
for idx, row in top_efficient.iterrows():
    print(f"  {row['District']}: {row['Overall_Yield_Efficiency']:.2f} MT/hectare")

print(f"\n df_preprocessed is ready for analysis!")
print(f"   Use this dataframe for all subsequent analyses to avoid reprocessing.")
```

```python


# Function to safely convert to numeric
def safe_numeric_convert(series):
    """Convert series to numeric, handling various string formats"""
    series = series.astype(str).str.replace(',', '')  # Remove commas
    series = series.str.replace('-', '0')  # Replace dashes with 0
    series = series.str.replace('nil', '0', case=False)  # Replace nil with 0
    series = series.str.replace('n/a', '0', case=False)  # Replace n/a with 0
    series = series.str.strip()  # Remove whitespace
    return pd.to_numeric(series, errors='coerce').fillna(0)

print("="*80)
print("YIELD EFFICIENCY HEATMAP ANALYSIS")
print("Objective: Identify which districts achieve highest yields for different crops")
print("="*80)

# Clean the dataframe
df_clean = df.copy()

# Get all production and area columns
production_cols = [col for col in df.columns if 'PROD' in col and 'M.T.' in col]
area_cols = [col for col in df.columns if 'AREA' in col and 'Hect.' in col]

# Convert all production and area columns to numeric
for col in production_cols + area_cols:
    if col in df_clean.columns:
        df_clean[col] = safe_numeric_convert(df_clean[col])

# Step 1: Find top 10 crops by total national production
crop_totals = {}
for prod_col in production_cols:
    if prod_col in df_clean.columns:
        # Extract crop name (remove 'PROD M.T.' part)
        crop_name = prod_col.replace(' PROD  M.T.', '').replace(' PROD. M.T.', '').replace(' PROD  M.T..1', '').strip()
        total_production = df_clean[prod_col].sum()
        crop_totals[crop_name] = {
            'total_production': total_production,
            'prod_col': prod_col
        }

# Sort and get top 10 crops
top_10_crops = sorted(crop_totals.items(), key=lambda x: x[1]['total_production'], reverse=True)[:10]

print("Top 10 Crops by National Production:")
for i, (crop, data) in enumerate(top_10_crops, 1):
    print(f"{i}. {crop}: {data['total_production']:.1f} MT")

# Step 2: Find corresponding area columns for these crops
crop_area_mapping = {}
for crop_name, crop_data in top_10_crops:
    prod_col = crop_data['prod_col']

    # Try to find matching area column
    possible_area_cols = [col for col in area_cols if crop_name.replace(' ', '') in col.replace(' ', '')]

    if not possible_area_cols:
        # Try more flexible matching
        crop_words = crop_name.split()
        for area_col in area_cols:
            if any(word in area_col for word in crop_words if len(word) > 3):
                possible_area_cols.append(area_col)
                break

    if possible_area_cols:
        area_col = possible_area_cols[0]  # Take the first match
        crop_area_mapping[crop_name] = {
            'prod_col': prod_col,
            'area_col': area_col
        }

print(f"\nFound area data for {len(crop_area_mapping)} out of 10 top crops")

# Step 3: Calculate yield for each crop-district pair
yield_data = []

for district in df_clean['District']:
    district_row = df_clean[df_clean['District'] == district].iloc[0]

    for crop_name, cols in crop_area_mapping.items():
        prod_col = cols['prod_col']
        area_col = cols['area_col']

        production = district_row[prod_col]
        area = district_row[area_col]

        # Calculate yield only if both area and production are > 0
        if area > 0 and production > 0:
            yield_value = production / area
            yield_data.append({
                'District': district,
                'Crop': crop_name,
                'Yield': yield_value,
                'Production': production,
                'Area': area
            })
        else:
            # Add as missing data point
            yield_data.append({
                'District': district,
                'Crop': crop_name,
                'Yield': np.nan,
                'Production': production,
                'Area': area
            })

# Convert to DataFrame
yield_df = pd.DataFrame(yield_data)

# Step 4: Create pivot table for heatmap
heatmap_data = yield_df.pivot(index='District', columns='Crop', values='Yield')

# Calculate average yield per district (for sorting)
district_avg_yield = heatmap_data.mean(axis=1, skipna=True).sort_values(ascending=False)

# Reorder districts by average yield
heatmap_data = heatmap_data.reindex(district_avg_yield.index)

# Step 5: Create the heatmap using matplotlib and seaborn
plt.figure(figsize=(14, 10))

# Create custom colormap (red to green)
cmap = sns.diverging_palette(10, 150, as_cmap=True)

# Create the heatmap
mask = heatmap_data.isnull()  # Mask for missing values

ax = sns.heatmap(
    heatmap_data,
    annot=True,  # Show values in cells
    fmt='.1f',   # Format to 1 decimal place
    cmap=cmap,
    center=heatmap_data.median().median(),  # Center colormap on median
    mask=mask,   # Gray out missing values
    cbar_kws={'label': 'Yield (MT per Hectare)'},
    linewidths=0.5,
    square=False
)

# Customize the plot
plt.title('Crop Yield Efficiency Heatmap by District\n(Higher values = more efficient production)',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Crops', fontsize=12, fontweight='bold')
plt.ylabel('Districts (sorted by average yield)', fontsize=12, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

# Extra Visualization for all the Crops

```python
import pandas as pd
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')

df_clean = df_preprocessed.copy()
print("="*60)
print("SUNBURST CHART - AGRICULTURAL PRODUCTION BREAKDOWN")
print("="*60)

# Define crop categories for sunburst
crop_categories = {
    'Grains': ['KURAKKAN PROD  M.T.', 'MAIZE PROD  M.T.', 'SORGHUM PROD  M.T.', 'MENERI PROD  M.T.'],
    'Vegetables': ['TOMATOES PROD  M.T.', 'RED ONIONS PROD  M.T.', 'CABBAGE PROD  M.T.',
                   'CHILLIES (GREEN) PROD  M.T.', 'BEANS PROD  M.T.', 'CUCUMBER PROD  M.T.',
                   'CARROT PROD  M.T.', 'BEETROOT PROD  M.T.'],
    'Root_Crops': ['MANIOC PROD  M.T.', 'SWEET POTATOES PROD  M.T.', 'POTATOES PROD  M.T.'],
    'Pulses': ['GREEN GRAM PROD  M.T.', 'BLACK GRAM PROD  M.T.', 'COWPEA PROD  M.T.', 'SOYA BEANS PROD  M.T.'],
    'Others': ['GINGELLY PROD  M.T.', 'GROND NUTS PROD  M.T.', 'MUSTARD PROD  M.T.']
}

# Create data for sunburst
sunburst_data = []

for district in df_clean['District']:
    district_row = df_clean[df_clean['District'] == district].iloc[0]

    for category, crops in crop_categories.items():
        category_total = 0
        for crop in crops:
            if crop in df_clean.columns:
                category_total += district_row[crop]

        if category_total > 0:  # Only add if there's production
            sunburst_data.append({
                'District': district,
                'Category': category,
                'Production': category_total
            })

# Convert to DataFrame
sunburst_df = pd.DataFrame(sunburst_data)

# Create sunburst chart
fig = px.sunburst(
    sunburst_df,
    path=['Category', 'District'],  # Category is inner ring, District is outer ring
    values='Production',
    title='Agricultural Production by Crop Category and District<br><sub>Inner ring: Crop Categories | Outer ring: Districts</sub>',
    color='Production',
    color_continuous_scale='Viridis',
    width=800,
    height=800
)

# Customize the chart
fig.update_traces(
    textinfo="label+percent parent",
    hovertemplate='<b>%{label}</b><br>Production: %{value:.1f} MT<br>Percentage: %{percentParent}<extra></extra>'
)

fig.update_layout(
    font_size=12,
    title_font_size=16
)

# Show the chart
fig.show()
```
